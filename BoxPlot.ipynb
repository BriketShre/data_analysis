```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_excel("C:/Users/prita/OneDrive/Desktop/Python Data Analyst/mat/expense_list.xlsx")

df = pd.DataFrame(data)
print(df)
group1 = df.groupby("Category")["Amount"].sum()
group2 = df.groupby("Description")["Amount"].sum()
plots = (df["Amount"], group1, group2)
plt.boxplot(plots)
plt.show()
```
