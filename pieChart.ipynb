```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 


brands = ["One Plus", "Apple", "Samsung", "Nokia","Redmi"]
x = [10,40,35,5,10]
exp = [0,0,0,0.1,0]
plt.pie(x, labels = brands, explode=exp, shadow=True, autopct="%.f", startangle=45)
plt.show()
```
