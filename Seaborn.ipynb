```python
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt
 

data = sns.load_dataset("tips")

gp = data.groupby("day").agg({"tip":"mean"})
sns.heatmap(gp)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("tips")

a = sns.FacetGrid(data, col = "smoker")
a.map(sns.barplot, "day","tip")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("tips")

sns.relplot(data = data, x = "tip",y = "total_bill" ,hue="sex", kind="scatter", col = "day", size = "smoker")
plt.show()
```

swarm plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("tips")

sns.swarmplot(data = data, x ="day", y ="total_bill", hue="sex", dodge=True)
plt.legend(bbox_to_anchor = (1,1))
plt.show()
```
