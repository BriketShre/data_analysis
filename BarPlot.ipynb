```python
import matplotlib.pyplot as plt

y = [98,88,97,93]
x = ["Part1","Part2","Part3","Part4"]
colours = ["Red","green", "blue","yellow"]
plt.bar(x ,y, color = colours)
plt.xlabel("Parts of Harry Potter", fontsize = 17)
plt.ylabel("Popularity", fontsize = 17)
plt.title("Popularity of Different Parts Of Haryy Potter", fontsize = 20)
plt.show()
```
